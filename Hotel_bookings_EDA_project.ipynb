## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df1.isnull().sum().sort_values(ascending = False)[:6]
````

````python
df1[['company','agent','children']] = df1[['company','agent','children']].fillna(0)
df1[['country']] = df1[['country']].fillna('other')
````

````python
# checking all null values is removed
df1.isnull().sum().sort_values(ascending = False)[:6]
````

````python
# checking the duplicate values in datasets
len(df1[df1.duplicated()])
````

````python
# Dropping the duplicates Rows
df1 = df1.drop_duplicates()
````

````python
# checking the shape of dataset whose combining values of adults,babies and childrens columns is 0
df1[df1['adults']+df1['babies']+df1['children'] == 0].shape
````

````python
df1.shape
````

````python
# Dropping the rows where combining the values of adults, childrens and babies columns is 0 because of no booking was made
df1.drop(df1[df1['adults']+df1['babies']+df1['children'] == 0].index, inplace = True)
````

### **Converting columns to appropriate datatypes.**

````python
# Converting datatype of columns 'children', 'company' and 'agent' from float to int.
df1[['children', 'company', 'agent']] = df1[['children', 'company', 'agent']].astype('int64')
````

````python
# changing datatype of column 'reservation_status_date' to data_type.
df1['reservation_status_date'] = pd.to_datetime(df1['reservation_status_date'])
````

### **Adding important columns.**

````python
# Adding total staying days in hotels
df1['total_stay'] = df1['stays_in_weekend_nights']+df1['stays_in_week_nights']

# Adding total people num as column, i.e. total people num = num of adults + children + babies
df1['total_people'] = df1['adults']+df1['children']+df1['babies']
````
